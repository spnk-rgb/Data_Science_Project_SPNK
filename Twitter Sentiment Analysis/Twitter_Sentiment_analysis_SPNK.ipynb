```python
def message_cleaning(message):
    test_punc_removed=[char for char in message if char not in string.punctuation]
    test_punc_removed_join=''.join(test_punc_removed)
    test_punc_removed_join_clean=[word for word in test_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
    return test_punc_removed_join_clean
```

```python
tweets_df_clean=tweets_df['tweet'].apply(message_cleaning)
```

```python
print(tweets_df_clean[5])
```

```python
print(tweets_df['tweet'][5])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
t_vectorizer=CountVectorizer(analyzer=message_cleaning)
```

```python
tweets_countvectorizer=CountVectorizer(analyzer=message_cleaning,dtype='uint8').fit_transform(tweets_df['tweet']).toarray()
```

```python
tweets_countvectorizer.shape
```

```python
l=tweets_countvectorizer
```

```python
m=tweets_df['label']
```

```python
l.shape
```

```python
m.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
l_train,l_test,m_train,m_test=train_test_split(l,m,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
NB_classifier=MultinomialNB()
```

```python
NB_classifier.fit(l_train,m_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
m_predict_test=NB_classifier.predict(l_test)
```

```python
cm=confusion_matrix(m_test,m_predict_test)
```

```python
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(m_test,m_predict_test))
```
